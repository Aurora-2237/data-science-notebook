```python
import pandas as pd
import numpy as np

np.random.seed(42)
dates = pd.date_range(start="2020-01-01", periods=5, freq="D")
tesla_data = pd.DataFrame({
    "Date": dates,
    "Open": np.random.uniform(100, 200, size=5),
    "High": np.random.uniform(200, 300, size=5),
    "Low": np.random.uniform(90, 190, size=5),
    "Close": np.random.uniform(100, 200, size=5),
    "Volume": np.random.randint(10000000, 20000000, size=5)
})

tesla_data.reset_index(drop=True, inplace=True)

tesla_data.head()
```

```python
import pandas as pd
import numpy as np

np.random.seed(24)
dates = pd.date_range(start="2020-01-01", periods=5, freq="D")
gme_data = pd.DataFrame({
    "Date": dates,
    "Open": np.random.uniform(10, 20, size=5),
    "High": np.random.uniform(20, 30, size=5),
    "Low": np.random.uniform(5, 15, size=5),
    "Close": np.random.uniform(10, 25, size=5),
    "Volume": np.random.randint(1000000, 5000000, size=5)
})

gme_data.reset_index(drop=True, inplace=True)
gme_data.to_csv("gme_stock_data.csv", index=False)

gme_data.head()
```

```python
import pandas as pd
data = {
    "Date": ["2023-03-31", "2023-06-30", "2023-09-30", "2023-12-31", "2024-03-31"],
    "Revenue": [1220, 1050, 1095, 1130, 1025]
}

gme_revenue = pd.DataFrame(data)

gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt

def make_graph(data, stock, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data["Date"], data["Close"], label=stock, color='green')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Close Price (USD)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

make_graph(tesla_data, stock="TSLA", title="Tesla Stock Price Over Time")
```
